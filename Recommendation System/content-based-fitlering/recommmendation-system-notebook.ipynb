Import pandas and read the CSV File

```python
import pandas as pd

df = pd.read_csv('dataset/BX-Books.csv',error_bad_lines=False,encoding='latin-1',sep=';')
df.head()
```

Show the columns' name

```python
df.info()
```

Count how many duplicated book title

```python
df.duplicated(subset='Book-Title').sum()
```

Remove every duplicated title

```python
df = df.drop_duplicates(subset='Book-Title')
```

Check the duplicated title again

```python
df.duplicated(subset='Book-Title').sum()
```

We take only 15k from the total 217k titles to minimize memory usage and runtime

```python
sample_size = 15000
df = df.sample(n=sample_size, replace=False, random_state=490)

df = df.reset_index()
df = df.drop('index',axis=1)
```

Check the dataframe again

```python
df.head()
```

Now we remove the whitespace of the authors name because it will affect the training if some authors have same words of their name

```python
def clean_text(author):
    result = str(author).lower()
    return(result.replace(' ',''))
```

```python
df['Book-Author'] = df['Book-Author'].apply(clean_text)
```

Check the author name

```python
df.head()
```

We also lowercase the title and the publisher

```python
df['Book-Title'] = df['Book-Title'].str.lower()
df['Publisher'] = df['Publisher'].str.lower()
```

We just want to use title, author, and publisher as the labels. So we remove other columns.

```python
df2 = df.drop(['ISBN','Image-URL-S','Image-URL-M','Image-URL-L','Year-Of-Publication'],axis=1)

df2['data'] = df2[df2.columns[1:]].apply(
    lambda x: ' '.join(x.dropna().astype(str)),
    axis=1
)

print(df2['data'].head())
```

We use sklearn to vectorize the data.

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
vectorized = vectorizer.fit_transform(df2['data'])
```

We use Cosine Similarity for our recommendation system. Cosine similarity ranges between 0 and 1. A value of 0 indicates that the two vectors are not similar at all, while 1 tells us that they are identical.

```python
from sklearn.metrics.pairwise import cosine_similarity

similarities = cosine_similarity(vectorized)
```

Print similarities variable.

```python
print(similarities)
```

Map the vector to previous dataframe. It should print value of 1 as diagonal beause the book is similar with itself.

```python
df = pd.DataFrame(similarities, columns=df['Book-Title'], index=df['Book-Title']).reset_index()

df.head()
```

Let's predict one book. It will print the most 10 similar book with the tested book.

```python
input_book = 'celestine insights'
recommendations = pd.DataFrame(df.nlargest(11,input_book)['Book-Title'])
recommendations = recommendations[recommendations['Book-Title']!=input_book]
print(recommendations)
```
